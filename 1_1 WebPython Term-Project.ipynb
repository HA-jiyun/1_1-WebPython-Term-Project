```python
NumberList = ["201606","201607","201608","201609","201610","201611","201612","201701","201702","201703","201704","201705","201706",
              "201707","201708","201709","201710","201711","201712","201801","201802","201803","201804","201805","201806","201807",
              "201809","201810","201811","201812","201901","201902","201903","201904","201905"]
YesBSlist = []
YesList = []

for i in range(0,len(NumberList)) :
    YesBSlist.append("예스" + NumberList[i] + ".xlsx")
    YesList.append("Yes" + NumberList[i])

for j in range(0,len(YesBSlist)):
    YesList[j] = pd.read_excel(YesBSlist[j])
    display(YesList[j].head(15)) #1~15위만 출력함에 유의한다.
```

### 교보문고

```python
NumberList = ["201606","201607","201608","201609","201610","201611","201612","201701","201702","201703","201704","201705","201706",
              "201707","201708","201709","201710","201711","201712","201801","201802","201803","201804","201805","201806","201807",
              "201809","201810","201811","201812","201901","201902","201903","201904","201905"]
KyoboBSlist = []
KyoboList = []

for i in range(0,len(NumberList)) :
    KyoboBSlist.append("교보" + NumberList[i] + ".xlsx")
    KyoboList.append("Kyobo" + NumberList[i])

for j in range(0,len(KyoboBSlist)):
    KyoboList[j] = pd.read_excel(KyoboBSlist[j])
    display(KyoboList[j].head(17)) #1~15위만 출력함에 유의한다.
```

## (4) 분석을 위한 데이터의 가공

### <1> 알라딘의 경우

현재 데이터 속에는 분석에 불필요한 정보가 상당히 많다. 따라서 그러한 정보들을 다음과 같은 사유로 제거해준다.
1. **ISBN(International Standard Book Number], 국제표준도서번호) 값들과 부가기호**는 책을 선호함에 있어서 불필요한 정보이므로 제거한다.
2. 마일리지를 위해 독자들이 불필요한 책을 읽는 경우가 드물기 때문에 **마일리지** 정보 역시 삭제한다.
3. **할인액**은 할인율을 통해 충분히 유추할 수 있으며, 판매가와 중복되는 상징성을 가지므로 제거한다.
4. **순번/순위**는 index 값을 통해 쉽게 알아낼 수 있는 정보이므로 삭제한다.
5. 데이터 선정 조건에서 국내도서로 한정해두었으므로, **구분** 또한 삭제해준다.

```python
for j in range(0,len(AladinBSlist)):
    AladinList[j].drop('순번/순위', axis=1, inplace=True)
    AladinList[j].drop('구분', axis=1, inplace=True)
    AladinList[j].drop('ISBN', axis=1, inplace=True)
    AladinList[j].drop('ISBN13', axis=1, inplace=True)
    AladinList[j].drop('부가기호', axis=1, inplace=True)
    AladinList[j].drop('마일리지', axis=1, inplace=True)
    AladinList[j].drop('할인액', axis=1, inplace=True)
    
    display(AladinList[j].head(15))
```

따라서 필요한 데이터만 남게끔 가공하는 코드가 완성되었다.<br>
여기서 문제는, 알라딘, yes24, 교보문고에서 제공하는 각각의 엑셀 파일에 포함된 정보의 종류가 다르다는 것이다.<br>
알라딘의 경우 위와 같으나, 필요한 정보를 뽑아내고 통계를 내기 위해서는 yes24와 교보문고에 해당되는 코드를 별도로 작성해주어야 한다.

### <2> yes24의 경우

알라딘의 경우와 동일하게, yes24의 데이터 중에서도 불필요한 정보들을 다음과 같은 사유로 제거해준다.
1. **ISBN(International Standard Book Number], 국제표준도서번호) 값들과 상품번호**는 책을 선호함에 있어서 불필요한 정보이므로 제거한다.
2. 마일리지를 위해 독자들이 불필요한 책을 읽는 경우가 드물기 때문에 **YES포인트** 역시 삭제한다.
3. **설명**은 책의 선택에 있어 필요한 정보이긴 하지만, 이 프로젝트에서 분석하고자 하는 정보는 아니므로 제외한다.
4. **순위**는 index 값을 통해 쉽게 알아낼 수 있는 정보이므로 삭제한다.

```python
for j in range(0,len(YesBSlist)):
    YesList[j].drop('순위', axis=1, inplace=True)
    YesList[j].drop('ISBN', axis=1, inplace=True)
    YesList[j].drop('상품번호', axis=1, inplace=True)
    YesList[j].drop('YES포인트', axis=1, inplace=True)
    YesList[j].drop('설명', axis=1, inplace=True)
    
    display(YesList[j].head(15))
```

### <3> 교보문고의 경우

두 서점의 경우와 동일하게, 교보문고의 데이터 중에서도 불필요한 정보들을 다음과 같은 사유로 제거해준다.
1. **ISBN(International Standard Book Number], 국제표준도서번호) 값**은 책을 선호함에 있어서 불필요한 정보이므로 제거한다.
2. **순위**는 index 값을 통해 쉽게 알아낼 수 있는 정보이므로 삭제한다.

여기서 주의할 점은, 교보문고가 배포한 엑셀 파일에 나머지 두 서점과 다른 부분이 많다는 것이다.<br>
index[1]이 불필요한 행이고, 열의 명칭이 정보명이 아니라는 점에 주의하며 코드를 수정한다.

```python
for j in range(0,len(KyoboBSlist)):
    KyoboList[j].drop('Unnamed: 1', axis=1, inplace=True)
    KyoboList[j].drop(KyoboList[j].index[1], inplace=True)
    
    KyoboList[j].reset_index(inplace=True)
    KyoboList[j].drop('index', axis=1, inplace=True)
    display(KyoboList[j].head(16))
```

* 순위 카테고리의 경우, 열 이름이 '교보문고 종합 월간_ yyyy년 mm월 분야 종합 베스트 셀러'와 같이 명명되어 있다. 따라서 열 이름이 파일마다 변화하는 관계로, 순위는 남겨둔 상태로 출력했다.

주변의 의견을 들어보니 실제로 구매할 때에는 크게 고려하지 않지만, 영향이 없지는 않을 것이라는 입장이 많았다.<br>

이외에도, 온라인 서점에서는 배송비 무료나 사은품 등의 행사를 수시로 진행한다.<br>
이런 요인들을 전부 고려하지 않은 채로 결론을 내리는 것은 안일한 행동이라고 생각하게 되었다.<br>
